***Grafico de dispersión***

```python
import matplotlib.pyplot as plt 
from math import pi, sin

valores= [0, pi/2, pi, 3*pi/2, 2*pi ]
func_sena=[]

valores_1 = []

for val in range(0,21):
    valores_1.append((val * pi) / 10 )

print(valores_1)

for val in valores_1:
    func_sena.append(sin(val))
    
plt.scatter(valores_1, func_sena)
plt.show()


```

***Personalización de graficos***


```python
frutas =["Manzanas", "Fresa", "uva", "pera", "mora"]
ventas_2023 = [120,90,150,80,110]
ventas_2024 = [100,100,130,95,80]

fig, axs = plt.subplots(2,1)
axs[0].bar(frutas, ventas_2023)
axs[1].pie(ventas_2024, labels=frutas , autopct="%d%%" )
plt.show()
```

```python
frutas =["Manzanas", "Fresa", "uva", "pera", "mora"]
ventas_2023 = [120,90,150,80,110]
ventas_2024 = [100,100,130,95,80]

fig, axs = plt.subplots(2,1)
axs[0].bar(frutas, ventas_2023)
axs[1].pie(ventas_2024, labels=frutas , autopct="%d%%" )
plt.show()
```

```python
x = [0,2,4,6,8]

y = []
y_1 = []
print(y)

for data in x:
    y.append(2 * data + 1)

for data in x:
    y_1.append(-1 * data + 4 )
 

fig, axs = plt.subplots(3,1)
axs[0].plot(x,y)
axs[1].plot(x,y_1)
axs[2].plot(x,y,y_1)
plt.show()
```
