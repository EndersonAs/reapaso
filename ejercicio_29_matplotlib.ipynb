***Libreria Matplotlib***

```python
from matplotlib import pyplot
```

***Grafico de linea***

```python
años = [2020,2022,2024]
ventas = [3000,5000,4000]
pyplot.plot(años,ventas)
pyplot.show()
```

***usandolo con el alias plt***

```python
import matplotlib.pyplot as plt 

años = [2020,2022,2024]
ventas = [3000,5000,4000]
ventas_a = [0,1000,2000,3000,4000,5000,6000]

plt.plot(años,ventas)
plt.xticks(años) #sirve para personalizar el formato de los ejes X y Y
plt.yticks(ventas_a) 
plt.show()

```

***Grafico de dispersión***

```python
import matplotlib.pyplot as plt 
from math import pi, sin

valores= [0, pi/2, pi, 3*pi/2, 2*pi ]
func_sena=[]

valores_1 = []

for val in range(0,21):
    valores_1.append((val * pi) / 10 )

print(valores_1)

for val in valores_1:
    func_sena.append(sin(val))
    
plt.scatter(valores_1, func_sena)
plt.show()


```

***Grafica de barras***

```python
import matplotlib.pyplot as plt 

producto = ["gaseosa", "agua", "monster", "redbull"]
precio = [3500,2500,8000,7000]

plt.bar(producto,precio)
plt.show()
```
